### Распределение оценок
Временное распределение: пик активности приходится на 2000 год (10,061 оценка), что может быть связано с ростом популярности онлайн-кинотеатров

```python
%%timeit -r1 -n1
ratings = Ratings('ml-latest-small/ratings.csv')
ratings_movies = ratings.Movies(ratings)
ratings_by_year = ratings_movies.dist_by_year()
print("Распределение оценок по годам:")
for year, count in list(ratings_by_year.items())[:5]:
    print(f"{year}: {count} оценок")
```

Распределение оценок неравномерно: наиболее частые оценки - 4.0 (26,818 раз) и 3.0 (20,047 раз)

```python
%%timeit -r1 -n1
ratings = Ratings('ml-latest-small/ratings.csv')
ratings_movies = ratings.Movies(ratings)
rating_dist = ratings_movies.dist_by_rating()
print("Распределение оценок:")
for rating, count in rating_dist.items():
    print(f"Оценка {rating}: {count} раз")
```

Популярность фильмов: фильмы 1990-х годов ("Forrest Gump", "Pulp Fiction") остаются самыми обсуждаемыми

```python
%%timeit -r1 -n1
ratings = Ratings('ml-latest-small/ratings.csv')
ratings_movies = ratings.Movies(ratings)
top_rated = ratings_movies.top_by_num_of_ratings(10)
print("Топ-10 самых обсуждаемых фильмов:")
for title, count in top_rated.items():
    print(f"{title}: {count} оценок")
```

```python
%%timeit -r1 -n1
ratings = Ratings('ml-latest-small/ratings.csv')
ratings_movies = ratings.Movies(ratings)
best_movies_avg = ratings_movies.top_by_ratings(10, 'average')
print("Топ-10 фильмов по средней оценке:")
for title, rating in best_movies_avg.items():
    print(f"{title}: {rating}")
```

### Поведение пользователей
Активность пользователей: пользователь ID 414 оставил 2,698 оценок - самый активный участник

```python
%%timeit -r1 -n1
ratings = Ratings('ml-latest-small/ratings.csv')
ratings_users = ratings.Users(ratings)
active_users = ratings_users.top_by_num_of_ratings(10)
print("Топ-10 пользователей с наибольшим числом оценок:")
for user_id, count in active_users.items():
    print(f"ID {user_id}: {count} оценок")
```


Щедрость оценок: пользователь ID 53 имеет среднюю оценку 5.0, что может указывать либо на низкую критичность, либо на тщательный выбор фильмов

```python
%%timeit -r1 -n1
ratings = Ratings('ml-latest-small/ratings.csv')
ratings_users = ratings.Users(ratings)
strict_users = ratings_users.top_by_ratings(10, 'average')
print("Топ-10 пользователей с наивысшей средней оценкой:")
for user_id, rating in strict_users.items():
    print(f"ID {user_id}: средняя оценка {rating}")
```

Консерватизм оценок: у многих пользователей медианная оценка составляет 5.0, что свидетельствует о преобладании высоких оценок

```python
%%timeit -r1 -n1
ratings = Ratings('ml-latest-small/ratings.csv')
ratings_users = ratings.Users(ratings)
strict_users = ratings_users.top_by_ratings(10, 'median')
print("Топ-10 пользователей с наивысшей медианной оценкой:")
for user_id, rating in strict_users.items():
    print(f"ID {user_id}: медиана {rating}")
```

```python
tags = Tags('ml-latest-small/tags.csv')
```

### Особенности тегов
Специфические теги: наличие тегов типа "the catholic church is the most corrupt organization in history" показывает эмоциональную вовлеченность пользователей

```python
%%timeit -r1 -n1
complex_tags = tags.most_words(5)
print("Теги с наибольшим количеством слов:")
for tag, words in complex_tags.items():
    print(f"'{tag}': {words} слов")
```

Длина тегов: встречаются очень длинные теги (до 85 символов) с большим количеством слов

```python
%%timeit -r1 -n1
longest_tags = tags.longest(5)
print("Самые длинные теги:")
for tag in longest_tags:
    print(f"'{tag}' ({len(tag)} символов)")
```

```python
%%timeit -r1 -n1
print(tags.most_words_and_longest(5))
```

Популярные теги: "In Netflix queue" используется 131 раз, что отражает современные привычки просмотра

```python
%%timeit -r1 -n1
popular_tags = tags.most_popular(10)
print("Топ-10 самых популярных тегов:")
for tag, count in popular_tags.items():
    print(f"'{tag}': {count} раз")
```

```python
%%timeit -r1 -n1
funny_tags = tags.tags_with('funny')
print(f"Найдено {len(funny_tags)} тегов со словом 'funny':")
for tag in funny_tags[:10]:
    print(f"  * {tag}")
```

**Class Links**

```python
links = Links('ml-latest-small/links.csv')
```

### Экономические показатели (IMDB)

```python
%%timeit -r1 -n1
movie_ids = links.read_csv_column('ml-latest-small/links.csv', 'imdbId')[:10]  
fields = ['Director', 'Budget', 'Cumulative Worldwide Gross', 'Runtime']

imdb_data = links.get_imdb(movie_ids, fields)
for data in imdb_data[:3]:
    print(data)
```

```python
%%timeit -r1 -n1
top_directors = links.top_directors(5)
print("Топ-5 самых продуктивных режиссеров:")
for director, count in top_directors.items():
    print(f"{director}: {count} фильмов")
```

Бюджеты: самый высокий бюджет у фильма с ID 0113497 ($65,000,000)

```python
%%timeit -r1 -n1
links_most_expensive = links.most_expensive(5)
print("Топ-5 самых большим бюджетом:")
for movie_id, budget in links_most_expensive.items():
    print(f"{movie_id}: {budget}")
```

Прибыльность: "Toy Story" (0114709) показал самую высокую прибыль ($371,157,969)

```python
%%timeit -r1 -n1
links_most_profitable = links.most_profitable(5)
print("Топ-5 самых прибыльных фильмов:")
for title, profits in links_most_profitable.items():
    print(f"{title}: {profits}")
```

```python
%%timeit -r1 -n1
links_longest = links.longest(5)
print("Топ 5 фильмов с самым большим хронометражом")
for title, runtimes in links_longest.items():
    print(f"{title}: {runtimes}")
```

Стоимость производства: фильм "Джуманджи" (0113497) имеет самую высокую стоимость минуты экранного времени ($625,000)

```python
%%timeit -r1 -n1
links_top_cost_per_minute = links.top_cost_per_minute(5)
print("Топ-5 фильмов с самым бюджетом на минуту")
for title, costs in links_top_cost_per_minute.items():
    print(f"{title}: {costs}")
```
